```python
while True:
    t = input('请输入三角形三个边长，用单空格隔开：').split(' ')
    if len(t) == 1 and t[0] == 'q':
        print('已退出')
        break
    if len(t) != 3:
        print('输入错误，请重新输入')
        continue
    try:
        a,b,c = t
        a,b,c = int(a),int(b),int(c)
    except:
        print('有非数字，输入错误，请重新输入')
        continue
    if a > 0 and b > 0 and c > 0:
        if a + b > c and b + c > a and a + c > b:
            if a == b == c:
                print('等边三角形')
            elif a == b or a == c or b == c:
                print('等腰三角形')
            else:
                print('普通三角形')
        else:
            print('非三角形')
    else: 
        print('非三角形')
```

```python
n = 100
for i in range(n):
    for j in range(n):
        if j < n // 2 and i < j:
            print('*',end = '')
        elif j >= n // 2 and i + j + 1 <= n:
            print('*',end = '')
        else:
            print(' ',end = '')
    print()
```
